```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import pandas as pd
import os


algo_list = ['aes','des3','blowfish','rc4','rsa']
for df_no in range(1,31):
    for algo in algo_list:
        for row in range(1,20):
            
        # Load the CSV file without headers
            
            csv_path = f"/home/christina/final-project/v2/csv/cluster_analysis/d{df_no}/{algo}/df{df_no}_{algo}_row{row}.csv"
            data = pd.read_csv(csv_path, header=None)  # Load the file without headers
            
            # Print the structure of the data to confirm
            print(data.head())
            print(data.columns)
            
            # Use the first two columns as features
            X = data.iloc[:, [0, 1]].values  # Select the first and second columns
            
            # Standardize features
            X = StandardScaler().fit_transform(X)
            
            # Compute DBSCAN
            db = DBSCAN(eps=0.3, min_samples=10).fit(X)
            labels = db.labels_
            
            # Number of clusters in labels, ignoring noise if present
            n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
            n_noise_ = list(labels).count(-1)
            
            # print(f'Estimated number of clusters: {n_clusters_}')
            # print(f'Estimated number of noise points: {n_noise_}')
            
            # Plotting the result
            unique_labels = set(labels)
            colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
            
            for k, col in zip(unique_labels, colors):
                if k == -1:
                    # Black used for noise.
                    col = [0, 0, 0, 1]
            
                class_member_mask = (labels == k)
            
                core_samples_mask = np.zeros_like(labels, dtype=bool)
                core_samples_mask[db.core_sample_indices_] = True
            
                xy = X[class_member_mask & core_samples_mask]
                plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col), markeredgecolor='k', markersize=14)
            
                xy = X[class_member_mask & ~core_samples_mask]
                plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col), markeredgecolor='k', markersize=6)
            
            plt.title(f'Estimated number of clusters: {n_clusters_}')
            print(df_no," ",row," ",algo," :",n_clusters_)
            os.chdir(f"/home/christina/final-project/v2/images/clustering/{algo}/d{df_no}"            
            plt.savefig(f'df{df_no}_{algo}_row{row}.png')
            plt.show()

# algo_name row_no d1_clusters d2_clusters 
#  aes       1       5           6           3       
#  aes       2 
#  rsa       1     
```
